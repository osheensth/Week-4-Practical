```python
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0)
```

Now step through the examples in the tutorial to learn a little about the plot methods available in Seaborn.  

```python
df.describe()
```

In the above table we have the information on Pokeman  with the feature like Attack, Defense, Speed and Heart Power.We have a data of around 151 Pokemen which are on categorized into different Stage. Looking into the Stage field there are only there stages 1,2 and 3. 

The pokemon in stage 3 level higest attack, Defense, Speed and Heart Power where as the pokemon in the stage 1 has low attack, defense, speed and heart power. 

The speed is another numeric fields that helps to determine the speed of the Pokemon. The fatest pokemons has speed of 140 whereas the lowest speed is 15. As per my understanding, Sp.Atk and Sp.Def means speed attack and speed defense.

Let's explore the Pokemen data using the Seaborn graphics. 

Let's visualize the data with the scatter plot.

```python
sns.lmplot(x='Attack', y='Defense', data =df)
```

The Attack field is directly proportional to the field defense . Here when Attack is increasing then the Defense is also increasing. Lets remove the regression line from the figure and add another field Stage in the graphs but in colors.

```python
sns.lmplot(x='Attack', y='Defense', data =df , fit_reg =False , hue ='Stage')
```

In the above scatter plot, different stages are indicated by different colors. The green indicates the Pokemon in Stage3, orange indicates Stage2 and blue indicates stage 1.  Majority of the Pokemon falls under Stage1 and Stage2. There are few pokemons which are on stage 3. 

As we can see that pokemon in stage 3 has the higher attack and defense rate. Where as the Pokemon in the stage 1 is distributed from range 0 to 120 but is densed in the range 40-60. The Pokemon which are in stage 2 is highly densed in range 60-80. 

Lets visualize the same data in the box plot. 

```python
plt.figure(figsize=(9,6))

stat_df =df.drop(['Total','Stage','Legendary'], axis=1)
sns.boxplot(data = stat_df)
```

The box plot provide the summary statistic. The middle line in the box is the median value of the column. The gap between bottom whiskher line and median indicates 50% of the data below the median whereas the gap betow the middle line and the end of the whisker is the 50% of the data above the median.  The dots are the outliers.

The data shown eailer from df.describe() is visualized better from the box plot. 

We have violin plot which is similar to box plot but it helps to provide the more information about the disbritution of the data set.

```python
plt.figure(figsize = (11,6))

sns.violinplot(x='Type 1',y='Attack',data= df)
```

```python
We can custom the color of the violet plot. 
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
plt.figure(figsize=(11,6)) # Set plot dimensions
sns.violinplot(x='Type 1', y='Attack', data=df, 
               palette=pkmn_type_colors) # Set color palette
```

In the above violin plot we can see that the distribution of attack for different types. The fairy type has broder distribtion in the median. Where 

The swarm plot helps to visualize the each data point in the diagram. This plot is better to use when there is less data.

```python
plt.figure(figsize=(11,6))
sns.swarmplot(x='Type 1',y='Attack',data = df,palette = pkmn_type_colors)
```

In the above swarm plot, there are only 2 pokemon which falls in Ice type in which one of the pokemon has attack value 60 and the other one has 80.

We can also combine the 2 plots into the single plot.

```python


plt.figure(figsize =(11,6))

sns.violinplot( x= 'Type 1', y='Attack', data= df,inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)
 

sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent

plt.title('Attack by Type')
```

We can apply the concept of pivoting in the sea born. We can combinme many columns into one using the melt. Let keep the column Name , Type 1 and Type2 . All other fileds will be melted (ie combined ) in one single column.

```python
melted_df = pd.melt(stat_df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name="Stat")

melted_df.head()
```

Now lets create a swarm plot for new data frame melted_df .

```python
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1')
```

This is an interesting plot but a little hard to interpret! The next version splits the different points out by colour and is a little better.

```python
# 1. Enlarge the plot
plt.figure(figsize=(10,6))
 
sns.swarmplot(x='Stat', 
              y='value', 
              data=melted_df, 
              hue='Type 1', 
              dodge=True, # 2. Separate points by hue
              palette=pkmn_type_colors) # 3. Use Pokemon palette
 
# 4. Adjust the y-axis
plt.ylim(0, 260)
 
# 5. Place legend to the right
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

Histograms allow you to plot the distributions of numeric variables.

```python
# Distribution Plot (a.k.a. Histogram)
plt.figure(figsize=(9,6))
sns.histplot(df, x='Attack',
             kde=True) # Plot the density curve too
```

Bar plots help you visualize the distributions of categorical variables.

```python
plt.figure(figsize=(9,6))
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

## Plot Size and Title

One thing you may want to do is to change the plot size which by default can be a bit small.  You should also add titles to your plots and make sure they have good X and Y axis labels.  Here's an example of how you can do this.  Play around with these settings to see their effect.  (Note that I get a warning about deprecated arguements when running this, that's ok).

You might also want to experiment with different figure styles to find something you like.  Here is the [Seaborn documentation](https://seaborn.pydata.org/archive/0.10/tutorial/aesthetics.html#seaborn-figure-styles) on that for version 0.10.

```python
plt.figure(figsize=(12,6))
with sns.axes_style('darkgrid'):
    sns.distplot(df.Speed)
plt.title("Distribution of Speed for Pokemon")
```
